```python
import pandas as pd
import numpy as np
import pickle
import json
```

```python
flights = 0
airports = ['ATL','CLT','DEN','DFW','EWR','IAH','JFK','LAS','LAX','MCO','MIA','ORD','PHX','SEA','SFO']
for i in range(1,13):
    p1 = (pd.read_csv('C:\\Users\\Jayaraman\\Desktop\\Flight project\\Flight data\\2016\\On_Time_On_Time_Performance_2016_'+str(i)+'\\On_Time_On_Time_Performance_2016_'+str(i)+'.csv',low_memory = False)[['Origin','Dest','FlightDate','Quarter','Year','Month','DayofMonth','DepTime','DepDel15','CRSDepTime','DepDelayMinutes','OriginAirportID','DestAirportID','ArrTime','CRSArrTime','ArrDel15','ArrDelayMinutes']])
    p2 = (pd.read_csv('C:\\Users\\Jayaraman\\Desktop\\Flight project\\Flight data\\2017\\On_Time_On_Time_Performance_2017_'+str(i)+'\\On_Time_On_Time_Performance_2017_'+str(i)+'.csv',low_memory = False)[['Origin','Dest','FlightDate','Quarter','Year','Month','DayofMonth','DepTime','DepDel15','CRSDepTime','DepDelayMinutes','OriginAirportID','DestAirportID','ArrTime','CRSArrTime','ArrDel15','ArrDelayMinutes']])
    p1 = p1.loc[p1['Origin'].isin(airports)]
    p1 = p1.loc[p1['Dest'].isin(airports)]
    p2 = p2.loc[p2['Origin'].isin(airports)]
    p2 = p2.loc[p2['Dest'].isin(airports)]
    p = pd.concat([p1,p2],ignore_index = True)
    if i==1:
        flights = p;
    else:
        flights = pd.concat([flights,p],ignore_index = True)

```

```python
flights.dropna()
flights['hour'] = flights['CRSDepTime']//100 + (flights['CRSDepTime']%100)//30
flights['hour'] = np.where(flights['hour']>=24,23,flights['hour'])
flights['primkey'] = flights.apply(lambda row :row['Origin']+ ' : ' +str(row['FlightDate']) + ' : '+str(row['hour']) ,axis = 1)
flights.to_csv("C:\\Users\\Jayaraman\\Desktop\\Flight project\\Flight data\\flightdata.csv",encoding='utf-8', index=False)
```

```python
airports = ['ATL','CLT','DEN','DFW','EWR','IAH','JFK','LAS','LAX','MCO','MIA','ORD','PHX','SEA','SFO']
temp = []
reqvalues = []
for x in airports:
    for y in ['2016','2017']:
        for i in range(1,13):
            js = pd.read_json("C:\\Users\\Jayaraman\\Desktop\\Flight project\\Weather data\\"+x+"\\"+y+"-"+str(i)+".json")
            df = pd.DataFrame(js['data']['weather'])
            df = df[['date','hourly']]
            df['airport'] = x
            temp.append(df)
weather = pd.concat(temp,ignore_index = True)
_ = weather.apply(lambda row: [reqvalues.append([row['date'], hour, row['airport']]) for hour in row.hourly], axis=1)
    
```

```python
#print(reqvalues[0][1])
# for x in reqvalues:
#     y = x[1]
#     y = { key: y[key] for key in weatherDetails }
# print(reqvalues)
weatherdf = pd.DataFrame(reqvalues,columns = weather.columns)

weatherdf = pd.concat([weatherdf.drop(['hourly'],axis = 1),weatherdf['hourly'].apply(pd.Series)], axis = 1)
weatherDetails = ['date','airport','windspeedKmph','winddirDegree','weatherCode','precipMM','visibility','pressure','cloudcover','DewPointF','WindGustKmph','tempF','WindChillF','humidity','time']
weatherdf = weatherdf[weatherDetails]

```

```python
weatherdf['hour'] = weatherdf.index
weatherdf["primkey"] = weatherdf.apply(lambda x : x['airport']+' : '+str(x['date'])+' : '+str(x['hour']%24),axis = 1)
weatherdf.drop('hour',axis = 1)
weatherdf.index = weatherdf['primkey']
flights.index = flights['primkey']


```
