```
X = np.array(df.drop('ArrDel15',axis = 1).drop('ArrDelayMinutes',axis = 1))
y = np.array(df['ArrDel15'])
print(len(X),len(y))
X_train , X_test , y_train , y_test = ms.train_test_split(X,y,test_size = 0.2)

```

# RandomForestClassifier


```
file =  open('drive/My Drive/randomforestclassifier.pickle','wb') 
clf = pickle.load(file)
```

```
y_pred = clf.predict(X_test)
# target = ['Delayed','Not Delayed']
print(classification_report(y_test,y_pred,target_names = ['Not Delayed','Delayed']))
```

# DecisionTreeClassifier

```
clf = tree.DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```
y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred,target_names = ['Not Delayed','Delayed']))
```

# LogisticRegression

```
clf = LogisticRegression()
clf.fit(X_train,y_train)
```

```
y_pred = clf.predict(X_test)
print(classification_report(y_test,y_pred,target_names = ['Not Delayed','Delayed']))
clf.score(X_test,y_test)
```

```
clf = LinearRegression()
clf.fit(X_train,y_train)
```

```
# y_pred = clf.predict(X_test)
clf.score(X_test,y_test)
```

# ExtraTreesClassifier

```
clf = ExtraTreesClassifier(n_estimators = 150,n_jobs = -1,verbose = 1)
clf.fit(X_train,y_train)
```
