```
import pandas as pd
import numpy as np
import sklearn.model_selection as ms
from sklearn.ensemble import RandomForestRegressor,ExtraTreesRegressor,GradientBoostingRegressor,AdaBoostRegressor,BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from google.colab import drive
drive.mount('/content/drive')
```

# Getting Data

```
df = pd.read_csv('drive/My Drive/finaldatapredict.csv')
df = df[df['ArrDel15'] == 1.0]
```

```


X = np.array(df.drop('ArrDelayMinutes',axis = 1))
y = np.array(df['ArrDelayMinutes'])
X_train,X_test,y_train,y_test = ms.train_test_split(X,y,test_size = 0.1)
```

# RandomForestRegressor

```
reg = RandomForestRegressor(n_estimators = 200,n_jobs = -1,verbose = 1)
reg.fit(X_train,y_train)
```

```
y_pred =reg.predict(X_test)
```

```
import math
print(mean_absolute_error(y_test,y_pred))
print(math.sqrt(mean_squared_error(y_test,y_pred)))
print(r2_score(y_test,y_pred))
```

# ExtraTreeRegressor

```
reg = ExtraTreesRegressor(n_estimators = 200,n_jobs = -1,verbose = True)
reg.fit(X_train,y_train)
```

# LinearRegression

```
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```
import math
y_pred = reg.predict(X_test)
print(mean_absolute_error(y_test,y_pred))
print(math.sqrt(mean_squared_error(y_test,y_pred)))
print(r2_score(y_test,y_pred))
```

# GradientBoostingRegressor

```
reg = GradientBoostingRegressor(n_estimators = 200,verbose = True)
reg.fit(X_train,y_train)
```

```
y_pred = reg.predict(X_test)
import math
print(mean_absolute_error(y_test,y_pred))
print(math.sqrt(mean_squared_error(y_test,y_pred)))
print(r2_score(y_test,y_pred))
```

# AdaBoostRegressor

```
reg = AdaBoostRegressor(n_estimators = 200)
reg.fit(X_train,y_train)
```

```
y_pred = reg.predict(X_test)
import math
print(mean_absolute_error(y_test,y_pred))
print(math.sqrt(mean_squared_error(y_test,y_pred)))
print(r2_score(y_test,y_pred))
```
